Example

```python
from chemlog.cli import classify_chebi
import sys
from io import StringIO

# چند پپتید تستی
peptides = [
    "GAS",        # Gly-Ala-Ser
    "ACD",        # Ala-Cys-Asp
    "GGGG",       # Glycine-rich  peptide
]

# Arguments for the classify-chebi command
args_list = [
    '--chebi-version', '239',
    '--strategy', 'algo',
    '--only-peptides',
    '--n-molecules', '1000'
]

print("Note: The classify_chebi command is a CLI entry point designed for batch processing.\nIt does not take individual peptide sequences as direct input for classification.\nIts output is printed to stdout/stderr.")

# The loop below will call the batch classifier for each peptide, which might not be intended.
# If you need to classify individual peptides programmatically, you should look for a dedicated API function within the chemlog library.
for seq in peptides:
    print("\n======================")
    print("Attempting to run batch classification for current loop (peptide:", seq, ")")

    # Redirect stdout and stderr to capture the CLI output
    old_stdout = sys.stdout
    old_stderr = sys.stderr
    redirected_stdout = StringIO()
    redirected_stderr = StringIO()
    sys.stdout = redirected_stdout
    sys.stderr = redirected_stderr

    try:
        # Call the Click command's main method with the arguments list.
        # standalone_mode=False prevents sys.exit() on completion/error.
        classify_chebi.main(args=args_list, standalone_mode=False)
        cli_output = redirected_stdout.getvalue()
        cli_errors = redirected_stderr.getvalue()
    except SystemExit as e:
        # Click sometimes still calls sys.exit() even with standalone_mode=False
        cli_output = redirected_stdout.getvalue()
        cli_errors = redirected_stderr.getvalue()
        print(f"Click command exited with code {e.code}\nOutput: {cli_output}\nErrors: {cli_errors}", file=old_stderr)
    finally:
        sys.stdout = old_stdout
        sys.stderr = old_stderr

    print("CLI Standard Output (captured):\n", cli_output)
    if cli_errors:
        print("CLI Standard Error (captured):\n", cli_errors)

    # The 'result' variable will not contain the classified data in a structured Python object.
    # It will contain the captured output from the CLI.
    print("Note: The 'result' variable (captured stdout) for peptide", seq, "is shown above.")
```
